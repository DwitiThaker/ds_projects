```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tinu = pd.read_excel('Data_Train.xlsx')
tinu.head()
```

```python
tinu.info()
```

```python
tinu['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
tinu['Airline'] = le.fit_transform(tinu['Airline'])
```

```python
tinu['Airline']
```

```python
tinu['Day'] = pd.DatetimeIndex(tinu['Date_of_Journey']).day
tinu['Month'] = pd.DatetimeIndex(tinu['Date_of_Journey']).month
tinu['Year'] = pd.DatetimeIndex(tinu['Date_of_Journey']).year
```

```python
tinu.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
tinu.head()
```

```python
tinu['Total_Stops'].unique()
```

```python
tinu.dropna(inplace=True)
```

```python
tinu.isnull().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder
stops = [['non-stop','1 stop','2 stops', '3 stops', '4 stops' ]]
oe = OrdinalEncoder(categories=stops)
tinu['Total_Stops'] = oe.fit_transform(tinu[['Total_Stops']])
```

```python
tinu['Total_Stops']
```

```python
tinu['Source'] = le.fit_transform(tinu['Source'])
tinu['Destination'] = le.fit_transform(tinu['Destination'])
tinu['Additional_Info'] = le.fit_transform(tinu['Additional_Info'])
```

```python
tinu['Dep_Hour'] = pd.DatetimeIndex(tinu['Dep_Time']).hour
tinu['Dep_Minute'] = pd.DatetimeIndex(tinu['Dep_Time']).minute
tinu.drop('Dep_Time',axis=1 ,inplace=True)  #formatting to numeric
```

```python
#to categorize
def flight_dep_time(X):
    if X > 0 and X<12 :
      return 'Morning'
    elif X>=12 and X<17 :
      return 'Afternoon'
    elif X>=17 and X<21:
      return 'Night'
    else:
      return 'Mid-Night'

tinu['flight_time'] = tinu['Dep_Hour'].apply(flight_dep_time)
tinu.drop('Dep_Minute', axis=1, inplace=True)
tinu.drop('Dep_Hour', axis=1, inplace=True)
```

```python
#Encoding
tinu['dep_time_encod'] = le.fit_transform(tinu['flight_time'])
tinu.drop('flight_time',  axis=1, inplace=True)
```

```python
tinu['Arr_Time_hour'] = pd.DatetimeIndex(tinu['Arrival_Time']).hour
tinu['Arr_Time_minute'] = pd.DatetimeIndex(tinu['Arrival_Time']).minute

def flight_arr_time(Y):
    if Y > 0 and Y<12 :
      return 'Morning'
    elif Y>=12 and Y<17 :
      return 'Afternoon'
    elif Y>=17 and Y<21:
      return 'Night'
    else:
      return 'Mid-Night'

tinu['Arrival_Time'] = tinu['Arr_Time_hour'].apply(flight_arr_time)
tinu.drop('Arr_Time_minute', axis=1, inplace=True)
tinu.drop('Arr_Time_hour', axis=1, inplace=True)
```

```python
#Encod arrival time
tinu['Arr_time_encod'] = le.fit_transform(tinu['Arrival_Time'])
tinu.drop('Arrival_Time',  axis=1, inplace=True)
```

```python
tinu['Duration']
```

```python
tinu[['Hours', 'Minutes']] = tinu['Duration'].str.extract(r'(?:(\d+)h)?\s*(?:(\d+)m)?')

# Convert the extracted hours and minutes to integers
tinu['Hours'] = tinu['Hours'].fillna(0).astype(int)
tinu['Minutes'] = tinu['Minutes'].fillna(0).astype(int)

# Calculate total duration in minutes
tinu['Total_Minutes'] = tinu['Hours'] * 60 + tinu['Minutes']
```

```python
tinu.drop(['Duration', 'dep_time_encod','Arr_time_encod', 'Route', 'Additional_Info'], axis=1, inplace=True)
```

```python
tinu.drop(['Hours','Minutes'], axis=1, inplace=True)
```

```python
tinu.head()
```

```python
x = tinu.drop('Price', axis=1)
y = tinu['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
acc = r2_score(y_test, pred)
```

```python
print("Accuracy: ", acc*100)
```
